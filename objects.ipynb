## Geometry game

```python
import geometry_game
from geometry_game import Point, Rectangle
```

```python
point1 = geometry_game.Point(x=10, y=20) # this is object initiated from class "Point"
print(type(point1)) # once it print out, it says the object "point1" belongs to library geometry_game, class "Point"
print(type(3)) # similar to this one, it says the object "3" belongs to class "int"
print(point1) # pintt out the onject created, 
```

```python
# Calculate distance from point 1 to the origin of corrdinate
point1.distance_from_origin()
```

```python
# Calculate distance from point 1 to the other point with x y
point1.distance_from_point(10, 5)
```

```python
# Calculate the distance from point 1 to point 2 as the object
point2 = geometry_game.Point(5,100)
point1.distance_from_point_object(point2)
```

```python
# Create the regtangle from 2 point
pointx = Point(6,7)
rectanglex = Rectangle(Point(5,6), Point(7,9))
## The rectanglex is the object that takes two objects lowleft = point1, upright = point1
print(rectanglex.lowleft, rectanglex.upright)
## We can get access to the attribute of rectanglex and lowleft as well
print(rectanglex.lowleft.x, rectanglex.upright.x)
```

```python
pointx.falls_in_rectangle(rectanglex)
```

```python
from geometry_game import Rectangle, Point
from random import randint
# Create a random rectangle of any size
rectangle = Rectangle(
    Point(randint(0,9), randint(0,9)),
    Point(randint(10,19), randint(10,19))
)
# Print out the message for coordinate of the random rectangle
print("Rectangle coordinates: ", rectangle.lowleft.x, ",", rectangle.lowleft.y, "and", rectangle.upright.x, ",", rectangle.upright.y)
# Input the point from user
user_point = Point(float(input("Guess x: ")), float(input("Guess y: ")))
print("Your point is: ", user_point.x, ",", user_point.y)
# Check the point is inside rectangle or not
print("Your point was inside rectangle: ", user_point.falls_in_rectangle(rectangle))
# Guess the area of the rectangle
user_area = float(input("Guess rectangle area: "))
print(f"Your guess is {user_area}, and the true area is {rectangle.area()}")
```
